```python
def matrices():
    a = []
    b = []

    for contador in range(2):
      while True:
        try:
          dimension = list(map(int, input("Ingrese la cantidad de filas y la de columnas separadas por coma de la matriz: ").split(",")))

          for i in range(dimension[0]):
            while True:
              try:
                fila = list(map(float, input(f"Ingrese los elementos de la fila {i + 1} separados por comas(debe tener {dimension[1]} valores): ").split(",")))

                if len(fila) != dimension[1]:
                  print(f"Error, debe ingresar {dimension[1]} valores")
                else:
                  if contador == 0:
                    a.append(fila)
                    break
                  else:
                    b.append(fila)
                    break
              except ValueError:
                print("Error: Debe ingresar numeros para los valores de las filas")
          break
        except ValueError:
          print("Error: Debe ingresar valores enteros para la cantidad de filas y columnas")

    print("Sus matrices son:")
    print("\nA:")
    for fila in a:
      print(fila)
    print("\nB:")
    for fila in b:
      print(fila)
    return a, b
```

```python
def suma():
  a, b = matrices()

  if len(a) != len(b) or len(a[0]) != len(b[0]):
    print("Error: Las matrices deben tener las mismas dimensiones para sumarse.")
  else:
    result = []
    for i in range(len(a)):
      fila = []
      for j in range(len(a[0])):
        fila.append(a[i][j] + b[i][j])
      result.append(fila)

    print("\nResultado(Recuerde que se suma componente a componente):")
    for fila in result:
      print(fila)
```

```python
def multiplicacion():
  a, b = matrices()

  if len(a[0]) != len(b):
    print("Error: la primera matriz debe tener igual cantidad de columnas que filas en la segunda matriz")

  else:
    resultado = []

    for i in range(len(a)):
      fila = []
      for j in range(len(b[0])):
        suma = 0
        for k in range(len(b)):
          suma += a[i][k] * b[k][j]

        fila.append(suma)
      resultado.append(fila)

    print("\nResultado(Recuerde que se hace el producto punto de cada fila de A con cada columna de B):")
    for fila in resultado:
      print(fila)
```

```python
def p_h():
  a, b = matrices()

  if len(a) != len(b) or len(a[0]) != len(b[0]):
    print("Error: Las matrices deben tener las mismas dimensiones")

  else:
    result = []

    for i in range(len(a)):
      fila = []
      for j in range(len(a[0])):
        fila.append(a[i][j] * b[i][j])
      result.append(fila)

    print("\nResultado(Recuerde que es la multiplicacion componente a componente):")
    for fila in result:
      print(fila)
```

```python
def p_k():
  a, b = matrices()

  result = []

  for i in range(len(a)):
    for j in range(len(b)):
      fila = []

      for k in range(len(a[0])):
        for l in range(len(b[0])):
          fila.append(a[i][k] * b[j][l])
      result.append(fila)

  print("\nResultado(Recuerde que es la multiplicacion de cada componente de A con toda la matriz B):")
  for fila in result:
    print(fila)
```

```python
print("Bienvenido a la calculadora de matrices")
operaciones = {
    "1": "Sumar Matrices",
    "2": "Multiplicar Matrices",
    "3": "Producto Hadamard",
    "4": "Producto Kronecker",
    "5": "Salir"
}
while True:
  try:
    opcion = int(input(f"¿Que quiere hacer?{operaciones}"))
    if opcion == 1:
      suma()
    elif opcion == 2:
      multiplicacion()
    elif opcion == 3:
      p_h()
    elif opcion == 4:
      p_k()
    elif opcion == 5:
      print("\nGracias por usar la calculadora de matrices")
      break
    else:
      print("Error: Ingrese una opcion valida")
  except ValueError:
    print("Error: Debe ingresar un numero")
```
