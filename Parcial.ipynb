Instrucción 1

```python
A = {1, 2, 3}
B = {1, 2, 4}
A.difference(B)
```

el metodo .difference() halla la interseccion entre los conjuntos seleccionados y esos elementos se los quita al conjunto al que se le aplica el metodo. En este caso, la interseccion es el conjunto {1, 2}, luego le quita estos elementos al conjunto A, quedando el conjunto {3}


```python
X = {1, 2, 3}
Y = {1, 2, 4}
X.symmetric_difference(Y)
```

el metodo .symmetric_difference() hace lo mismo que el metodo .difference() pero para la union de los cojuntos. En este caso, la interseccion seria el conjunto {1,2}, la union {1, 2, 3, 4} y si a los elementos del conjunto formado por la union le quitamos los de la interseccion nos da el conjunto {3, 4}

Instruccion 2

```python
A = {5, 6, 7, 8}
B = {5, 6}

A.issuperset(B)
```

el metodo .issuperset() revisa si el conjunto dentro del parentesis es subconjunto del conjunto al que se le aplico. En este caso, como B esta contenido en A, significa que B es subconjunto de A y por eso imprime el elemento booleano True.

```python
X = {11, 12, 13, 14}
Y = {11, 12}
X.isdisjoint(Y)
```

el metodo .isdisjoint() revisa si al conjunto al que se le aplica no esta contenido por el conjunto dentro del parentesis. Es decir, revisa si los conjuntos no comparten algun elemento. En este caso, como si comparten, imprimira el elemento booleano False.

Instrucción 3

```python
datos2 = {
    "a" : 1,
    "b" : 2
}
name = {
    "Nicolas" : 17,
    "Marian" : 18,
}
datos2.update(name)
print(datos2)
```

el metodo .update() "actualiza" el diccionario. Es decir, si una pareja clave:valor no esta en el diccionario, la agrega, pero si ya está entonces la actualiza, modificando el valor anterior por el ingresado.

```python
x = (("a", 1), ("b", 5))
datos2.update(x)
print(datos2)
```

```python
y = ["c", 1]
datos2.update(y)
```

Aca podemos observar 3 cosas:

1.   Para actualizar/modificar el diccionario con el metodo .update() necesitamos una tupla de tuplas o lista de listas o tupla de listas o lista de tuplas con al menos 2 elementos dentro de ellas que contengan un par clave:valor.
2.   Se puede modificar el valor de alguna pareja clave:valor insertando uno de los elementos dicho antes con una pareja repetida y una pareja con misma clave pero con valor distinto.
3. Si insertamos una pareja clave:valor repetida no sucede nada, a menos que sea el unico elemento que metamos, ya que, como dije en el 1. se necesitan al menos 2 parejas.

```python
def f(m):
  if isinstance(m, int) and m >= 0:
    for i in range(0, (m+1)):
      if i == 0 or i == 1:
        print(f"el numero es {i}")

      else:
        for j in range(2, i):
            if i % j == 0:
              print(f"{i} es igual a {j} * {i//j}")
              break
            else:
              print(f"{i} es un numero primo")
              break
  elif isinstance(m, str):
    print(f"{m} es una cadena, por favor ingrese un numero entero")
  elif isinstance(m, float):
    print(f"{m} es un numero racional, por favor, ingrese un numero entero")
  elif m < 0:
    print(f"{m} no es natural, ingrese un numero natural")
```

```python
m = 6
f(m)
```

```python
m = -3
f(m)
```

```python
m = 3.8
f(m)
```

El codigo revisa que tipo de elemento es, pidiendo que sea un numero natural si no lo es. Si es natural, revisa los numeros del 0 a ese numero y, si es 0 o 1 lo imprime, si es primo imrpime esto y si no lo es entonces muestra una pequeña factorizacion.

```python
def contador(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  numeros = "0123456789"

  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0
  contador5 = 0

  for letra in palabra:
    if letra in vocales:
        contador +=1
    elif letra in vocales.upper():
        contador1 +=1
    elif letra in consonantes:
        contador2 +=1
    elif letra in consonantes.upper():
        contador3 +=1
    elif letra in numeros:
      contador4 +=1
    else:
      contador5 +=1

  return {"Cantidad de vocales: " : contador + contador1,
          "Cantidad de vocales minisculas: " : contador,
          "Cantidad de vocales mayusculas: " : contador1,
          "Cantidad de consonantes: " : contador2 + contador3,
          "Cantidad de consonantes minusculas: " : contador2,
          "Cantidad de consonantes mayusculas: " : contador3,
          "Cantidad de numeros: " : contador4,
          "Cantidad de caracteres especiales: " : contador5
          }

contador("HolaMundO123@. ")

```

Explora cada letra de la palabra ingresada: si esta en vocales aumenta el contador, si no mira si es una vocal mayuscula, si no mira si es una consonante y asi hasta que termina la palabra y muestra el resultado.
