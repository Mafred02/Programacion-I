```python
print(set())

print(set([3, 2, 5, 8]))
print(set((5, 1.2, 3)))
print(set("3.141592"))

conjunto = set([2, 3, 3, 4])
conjunto1 = set("3.141592")

print(conjunto)

conjunto.add(5)
print(conjunto)

conjunto.remove(3)
print(conjunto)

print(conjunto.intersection(conjunto1))
print(conjunto1.issubset(conjunto))
```

```python
diccionario = {3: "Uno", 4: "Dos"}
diccionario[3] = "Tres"
print(diccionario)

dict_lista_tuplas = dict([(5, "dos"), (3, "uno"), (3, "Tres")])
print(dict_lista_tuplas)

dict_lista_string = dict(Uno = 1, Dos = 2, Tres = 3)
print(dict_lista_string)

dict_tipos = {5: "integer",
              3.1: "float",
              "texto" : "string",
              (5, 1): "tupla"}

print(dict_tipos)

dict_repeticion = {1: "Primero", 1: "Último"}
print(dict_repeticion)

print(diccionario,
      diccionario.keys(),
      diccionario.values(),
      diccionario.items())

claves = diccionario.values()
print(claves)
diccionario[1] = "One"
print(claves)
diccionario.pop(3)
print(diccionario)
```
